```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

from mtcnn import MTCNN
from scipy.spatial.distance import cosine

from openface.architecture import OpenFaceArchitecture
from facenet.architecture import InceptionResNetV2 as FaceNetArchitecture

from typing import Tuple
```

```python
def get_face(img: np.ndarray, box: Tuple[int, int, int, int]):
    """Crop image to a given face area.
    
    Parameter
    -----
    img: np.ndarray
        Image containing face(s)
    box: [x:int, y:int, w:int, h:int]
        Top-left point (x, y) of face area, width (w) x height (h)
        
    Return
    -----
    face: np.ndarray
    pt1: (int, int)
        Top-left point of face area
    pt2: (int, int)
        Bottom-right point of face area
    """
    x1, y1, width, height = box
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 + height
    face = img.copy()[y1:y2, x1:x2]
    return face, (x1, y1), (x2, y2)


def normalize(img:np.ndarray) -> np.ndarray:
    """Zero mean and unit variance normalizarion"""
    mean, std = img.mean(), img.std()
    return (img - mean) / std


def preprocess_face(face:np.ndarray, size:Tuple[int,int]) -> np.ndarray:
    """Preprocess face image for model to extract to feature vector."""
    face = normalize(face)
    face = cv2.resize(face, size)
    return face


def extract_feature(face:np.ndarray, face_encoder) -> np.ndarray:
    return face_encoder.predict(np.expand_dims(face, axis=0))[0]


def similarity(face1:np.ndarray, face2:np.ndarray):
    return 1 - cosine(face1, face2)
```

```python
detector = MTCNN()

def detect_faces(img:np.ndarray, return_results:bool=False):
    results = detector.detect_faces(img)
    faces = list()
    for res in results:
        face, _, _ = get_face(img, res["box"])
        faces.append(face)
        
    if return_results:
        return faces, results
    
    else:
        return faces
    
    
def extract_face_feat(face:np.ndarray, face_encoder, required_size:Tuple[int,int]):
    face = preprocess_face(face, size=required_size)
    return face_encoder.predict(np.expand_dims(face, axis=0))[0]
```

```python
facenet = FaceNetArchitecture()
facenet_weights_path = "facenet/facenet_keras_weights.h5"
facenet.load_weights(facenet_weights_path)

print("Input: ", facenet.input_shape)
print("Output: ", facenet.output_shape)
```

```python
openface = OpenFaceArchitecture()
openface_weights_path = "openface/openface_keras_weights.h5"
openface.load_weights(openface_weights_path)

print("Input: ", openface.input_shape)
print("Output: ", openface.output_shape)
```

```python
img_prayut1 = cv2.cvtColor(cv2.imread("prayut1.jpg"), cv2.COLOR_BGR2RGB)
img_prayut2 = cv2.cvtColor(cv2.imread("prayut2.jpg"), cv2.COLOR_BGR2RGB)
img_prawit1 = cv2.cvtColor(cv2.imread("prawit1.jpg"), cv2.COLOR_BGR2RGB)

faces = list()

for img in [img_prayut1, img_prayut2, img_prawit1]:
    faces_ = detect_faces(img)
    faces += faces_
```

```python
fig, axes = plt.subplots(1, len(faces))

for face, ax in zip(faces, axes.ravel()):
    ax.imshow(face)
```

```python
face_feats_openface = list()
face_feats_facenet = list()

for face in faces:
    face_feats_openface.append(extract_face_feat(face, openface, required_size=(96,96)))
    face_feats_facenet.append(extract_face_feat(face, facenet, required_size=(160,160)))
```

```python
from itertools import combinations_with_replacement
```

```python
for a, b in combinations_with_replacement(range(len(faces)), 3):
    print(a, b)
```

```python
print("OpenFace")

for face1_idx, face2_idx in combinations_with_replacement(range(len(faces)), 2):
    score = similarity(face_feats_openface[face1_idx], face_feats_openface[face2_idx])
    print(f"{face1_idx} vs {face2_idx}: {score}")
```

```python
print("FaceNet")

for face1_idx, face2_idx in combinations_with_replacement(range(len(faces)), 2):
    score = similarity(face_feats_facenet[face1_idx], face_feats_facenet[face2_idx])
    print(f"{face1_idx} vs {face2_idx}: {score}")
```
